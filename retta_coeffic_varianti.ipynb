```
import matplotlib.pyplot as pl
lx=[]
ly=[]
for x in range(-250,251):
    y=0.5*x+10
    lx.append(x)
    ly.append(y)
pl.plot(lx,ly)
#pl.grid(True)
#pl.show()
a=0.5
b=-1
c=10
learn_speed=0.06
for k in range(0,2):
  lx=[]
  ly=[]
  a=a+2*learn_speed
  b=b+2*learn_speed
  c=c+2+learn_speed
  for x in range(-250,251):
    y=(a*0.5*x+c)/(-b)
    lx.append(x)
    ly.append(y)
    pl.plot(lx,ly)
pl.grid(True)
pl.show()
```
